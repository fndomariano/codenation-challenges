```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
pd.set_option("display.max_rows", 200)
pd.set_option('display.max_columns', 500)
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
print(train.shape)
print(test.shape)
```

```python
train.corr()['IN_TREINEIRO'].sort_values(ascending=False)
```

```python
columns = [
    'NU_INSCRICAO', 'IN_TREINEIRO', 'TP_ST_CONCLUSAO', 'TP_PRESENCA_CN', 
    'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT'
]
```

```python
train = train[columns[1:7]]
```

```python
test['IN_TREINEIRO'] = None
test = test[columns]
```

```python
test.isna().sum()
```

```python
train.head()
```

```python
test.head()
```

```python
y_train = train['IN_TREINEIRO']
x_train = train[columns[2:7]]
x_test = test[columns[2:7]]
```

```python
y_train
```

```python
x_train.columns
```

```python
x_test.columns
```

```python
model = RandomForestClassifier(
    n_estimators=2000, 
    n_jobs=-1, 
    max_depth=100, 
    min_samples_leaf=4,
    min_samples_split=10, 
    random_state=0
)
```

```python
model.fit(x_train, y_train)
```

```python
print('Accuracy = {}'.format(model.score(x_train, y_train).round(2)))
```

```python
y_predict = model.predict(x_test)
```

```python
answer = test
answer['IN_TREINEIRO'] = y_predict
```

```python
answer.drop(columns=columns[2:7], inplace=True)
```

```python
answer.head()
```

```python
answer.to_csv('answer.csv',columns=['NU_INSCRICAO', 'IN_TREINEIRO'], sep=',', index=False)
```
